IMPORTING THE DEPENDENCIES

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

DATA VISUALISATION

```python
sns.set()
```

CREATING A VECTOR

```python
plt.quiver(0,0,4,5)
```

PLOTTING VECTOR IN GRAPH

```python
plt.quiver(0,0,4,5 , scale_units ='xy', angles='xy', scale = 1, color='r' )
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5 , scale_units ='xy', angles='xy', scale = 1, color='r' )
plt.quiver(0,0,-3,-6 , scale_units ='xy', angles='xy', scale = 1, color='y' )
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

VECTOR ADDITION

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = np.asarray([0,0,3,-2])

sum = vector_1 + vector_2

print(sum)
```

GRAPHICAL REPRESENTATION OF VECTOR ADDITION

```python
plt.quiver(0,0,2,3 , scale_units ='xy', angles='xy', scale = 1, color='r' )
plt.quiver(0,0,3,-2 , scale_units ='xy', angles='xy', scale = 1, color='y' )
plt.quiver(0,0,5,1 , scale_units ='xy', angles='xy', scale = 1, color='y' )
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

VECTOR SUBTRACTION

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = np.asarray([0,0,3,-2])

difference = vector_1 - vector_2

print(difference)
```

GRAPHICAL REPRESENTATION OF VECTOR SUBTRACTION

```python
plt.quiver(0,0,2,3 , scale_units ='xy', angles='xy', scale = 1, color='r' )
plt.quiver(0,0,3,-2 , scale_units ='xy', angles='xy', scale = 1, color='y' )
plt.quiver(0,0,-1,5 , scale_units ='xy', angles='xy', scale = 1, color='b' )
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

MULTIPLYING INDIVIDUAL VECTOR BY A SCALAR

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = 2*vector_1

print(vector_2)
```

GRAPHICAL REPRESENTATION

```python
plt.quiver(0,0,2,3 , scale_units ='xy', angles='xy', scale = 1, color='r' )
plt.quiver(0,0,4,6 , scale_units ='xy', angles='xy', scale = 1, color='b' )
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
